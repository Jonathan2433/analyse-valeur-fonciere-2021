```python
import geopandas as gp
import matplotlib.pyplot as plt
import pandas as pd
```

## Ouverture de mon dataset departement

```python
dpt_fr = gp.read_file('DATA/departements/departements-20180101.shp')
dpt_fr
```

## Suppression des DOM TOM

```python
dpt_fr = dpt_fr.drop(dpt_fr[(dpt_fr.nom == 'La Réunion') | (dpt_fr.nom == 'Martinique') | (dpt_fr.nom == 'Guadeloupe') | (dpt_fr.nom == 'Guyane') | (dpt_fr.nom == 'Mayotte')].index)
dpt_fr
```

## PLOT du dataset departement

```python
fig, ax = plt.subplots(figsize=(10, 8))
dpt_fr.plot(ax=ax,color='midnightblue')
plt.show()
```

## Ouverture du dataset valeur fonciere

```python
vf = pd.read_csv('DATA/valeursfoncieres-2021.txt', sep="|", low_memory=False)
vf
```

## Merge des 2 Dataset Departement et valeur fonciere

```python
df = pd.merge(
    vf,
    dpt_fr,
    how="left",
    left_on='Code departement',
    right_on='code_insee'
)
df
```

## Preprocessing

```python
df = df.loc[(df["Code type local"] < 3)]
df = df.loc[(df['No disposition'] == 1)]
df = df.loc[(df['Nature mutation'] == 'Vente')]
df = df.loc[(df['Nombre pieces principales'] > 0)]
df
```

```python
df = df.drop(columns=[
    'Date mutation',
    'No disposition',
    'Code service sages',
    'Reference document',
    '1 Articles CGI',
    '2 Articles CGI',
    '3 Articles CGI',
    '4 Articles CGI',
    '5 Articles CGI',
    'Nature mutation',
    'B/T/Q',
    'Code voie',
    'Code commune',
    'Prefixe de section',
    'No Volume',
    'Identifiant local',
    'Section',
    'No plan',
    'Nature culture speciale',
    'Code type local',
    'Nature culture',
    'Code postal',
    'No voie',
    'Type de voie',
    'Voie',
    'Commune',
    '1er lot',
    'Surface Carrez du 1er lot',
    '2eme lot',
    'Surface Carrez du 2eme lot',
    '3eme lot',
    'Surface Carrez du 3eme lot',
    '4eme lot',
    'Surface Carrez du 4eme lot',
    '5eme lot',
    'Surface Carrez du 5eme lot',
    'Nombre de lots',
    'Type local',
    'Surface terrain',
    'code_insee',
    'nuts3',
    'wikipedia',
    'surf_km2',
    'Nombre pieces principales',
    'nom'
])
df
```

```python
df['Valeur fonciere'] = df['Valeur fonciere'].astype("string").str.replace(',', '.').astype(float)
df['Surface reelle bati'] = df['Surface reelle bati'].astype(float)
df
```

```python
df['Prix au m²'] = df['Valeur fonciere'] / df['Surface reelle bati']
df
```

```python
df_price_m2 = df.groupby(['Code departement'])['Prix au m²'].mean()
df_price_m2
```

```python
df_price_m2_polygon = pd.merge(
    df_price_m2,
    dpt_fr,
    how="right",
    left_on='Code departement',
    right_on='code_insee'
)
df_price_m2_polygon
```

```python
df_price_m2_polygon = df_price_m2_polygon[['code_insee', 'nom', 'nuts3', 'wikipedia', 'surf_km2', 'geometry', 'Prix au m²']]
df_price_m2_polygon
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
df_price_m2_polygon.plot(ax=ax,color='midnightblue')
plt.show()
```
